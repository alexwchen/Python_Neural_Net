```python
num_mat = io.loadmat('ex3weights.mat')
theta1 = num_mat['Theta1']
theta2 = num_mat['Theta2']
```

## 1.4 FeedForward Algorithm
### -   Theta1 has size 25 * 401
### -   Theta2 has size 10 * 26

```python
def sigmoid (t):
    return 1/(1+exp(-t))

Xones = np.ones((5000,1))

_X = np.hstack((Xones, X))

a1 = sigmoid( dot(_X,theta1.T) )

_a1 = np.hstack((Xones, a1))


a2 = sigmoid( dot(_a1,theta2.T) )

out = a2.argmax(axis=1) + 1

acc = y==out.reshape(5000,1)
print float(sum(acc))/5000


```

## 1.5 Cost Function without Regularization

```python
a3 = a2
cost = 0
for i in range(10):
    _y = y==(i+1)
    c = -(_y * (log(a3)[:,i].reshape(5000,1)) ) -( (1-_y) * (log(1-a3)[:,i].reshape(5000,1)) )
    cost = cost + sum(c)
    
print cost/5000
```

## 1.6 Cost Function with Regularization

```python
a3 = a2
cost = 0
for i in range(10):
    _y = y==(i+1)
    c = -(_y * (log(a3)[:,i].reshape(5000,1)) ) -( (1-_y) * (log(1-a3)[:,i].reshape(5000,1)) )
    cost = cost + sum(c)
    
r = ( sum(pow(theta1[:,1:],2)) + sum(pow(theta2[:,1:],2)) )/2/5000

print cost/5000 + r

```
